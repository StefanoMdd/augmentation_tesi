```python
import pandas as pd
import numpy as np
```

```python
#load file csv
df = pd.read_csv("/home/stefano/projects/augmentation_tesi/data_utilities/final_labels_oasis-1.csv")
df.head()
```

```python
#selezione istanze con label "Cognitively normal"
id_sesMR_CN = np.array(df[df["dx1"] == "Cognitively normal"].list1_id)
print(" Ci sono", len(id_sesMR_CN), "immagini MRI di soggetti 'cognitively normal'")

#relative path "Cognitively normal" ("sub-" + id_soggetto + id_ses_mr)
relativePath_mr_CN = ["sub-" + i.split("_")[0] + "/ses-" + i.split("_")[2]  for i in id_sesMR_CN]
```

```python
#salvataggio relative path MRI "Cognitively normal" in file csv
pd.DataFrame(relativePath_mr_CN).to_csv("/home/stefano/projects/augmentation_tesi/data_utilities/relativePath_mr_CN.csv",index=False)
```
